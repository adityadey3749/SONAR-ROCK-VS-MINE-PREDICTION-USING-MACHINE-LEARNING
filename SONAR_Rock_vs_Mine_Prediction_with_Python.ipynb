# SONAR ROCK VS MINE PREDICTION SYSTEM

## IMPORTING THE DEPENDENCIES

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

##DATA COLLECTION AND DATA PROCESSING

```python
#LOADING THE DATASETS TO PANDAS DATAFRAMES
sd = pd.read_csv("/content/SONAR DATASET.csv", header=None)
```

```python
sd.tail()
```

```python
#NUMBER OF ROWS AND COLUMNS
sd.shape
```

```python
sd.describe()
```

```python
sd[60].value_counts()
```

```python
#  GROUPING DATA BASED ON MINE AND ROCK

sd.groupby(60).mean()
```

```python
sd.head()
```

```python
# Seperating Data and Labels

X = sd.drop(columns=60, axis=1)
Y = sd[60]
```

```python
print(X)
print(Y)
```

```python
# Training and Test Data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)

# X-train     -   We will train all the 208 rows and 60 columns according to the Model
# X_test      -   We will test all those above columns
# Y_train     -   We will train all the 208 rows and only the last 1 column to the model
# Y-test      -   We will test that particular column above

# X, Y        -   Loading of Dataset X and Y

# test_size   -   Says about how much % we want to test 0.1 means 10% test data

# startify    -   Means which X or Y model is to be chosen(Splitting of data based on Rock and Mine)

# ranadom_state   - Reproduce the code as it is.
```

```python
print(X.shape)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

# Model Training
## Logistic Regression Model

```python
model = LogisticRegression()
```

```python
#Training the  Logistic Regression Model with training data

model.fit(X_train, Y_train)
```

```python
# Model Evaluation

#Accuracy Score Test
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy Score", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy Score", testing_data_accuracy)
```
